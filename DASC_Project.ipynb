#Analysis of Buildings and Conditions Report

Housing properties were rated from 1 to 4 in the 2018 survey for CDBG grant.

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px

%matplotlib inline

plt.rcParams['figure.figsize'] = [4, 3]
plt.rcParams['figure.dpi'] = 150
```

```
sheet_url = 'https://docs.google.com/spreadsheets/d/1SEI9P_qNzOOQj2uwO6wEsTLjp_TIHlpg0kpqWHeRNfo/edit#gid=924858711'
csv_export_url = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')

df = pd.read_csv(csv_export_url)

df.head()
```

## Data Visualization

The scale used to rate survey data in 2018: 

**1** - *Excellent*, **2** - *Good*, **3** - *Fair*, **4** - *Poor*

```
outcomes = df.columns[2:-1]
outcomes

for p in outcomes: 
  plt.hist(df[p], color='#2e4053', edgecolor='grey', bins=np.arange(6)-0.5)
  plt.xlabel("Score")
  plt.ylabel('Number of Properties')
  plt.xticks(range(1, 5))
  plt.title(f"Histogram of {p} Ratings")
  plt.show()
```
